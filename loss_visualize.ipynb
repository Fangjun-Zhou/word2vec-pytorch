```python
LOSS_PATH = "models/skipgram/loss.json"
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read loss data
lossData = json.loads(open(LOSS_PATH).read())
epochs = len(lossData["train"])
epochs
```

```python
# Plot train and validation loss over epochs.
plt.figure()
plt.plot(np.arange(0, epochs), lossData["train"], label="train_loss")
plt.plot(np.arange(0, epochs), lossData["val"], label="val_loss")
```
